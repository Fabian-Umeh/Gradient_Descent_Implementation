```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

Univariate batch gradient descent

```python
def Gradient_descent(X,y,epoch,lr):
  w = 1          #np.ones(X.shape[1]) number of feature
  b = 0

  costs = []
  epochs = []
  n= len(y)

  for i in range(epoch):
    yhat = b + (w * X.T)
                                      #getting the mean of the SSR
    grad_b = (-2/n) * np.sum(y - yhat)
    grad_w = (-2/n) * np.sum(X * (y - yhat))

    w = w -  lr * grad_b
    b = b -  lr * grad_w

    cost = np.mean((y - (yhat))**2)
    epochs.append(i)
    costs.append(cost)

  return costs, w, b, epochs
```

Multivariate Gradient descent

```python
def GD_multi(X,y,epoch, lr):

  w = np.ones(X.T.shape[1])
  b = 0
  costs = []
  epochs = []

  n = len(X)
  for i in range(epoch):
    yhat = b + np.dot(w, X)
    grad_b = (-2/n) * np.sum(y - yhat)
    grad_w = (-2/n) * (X.dot(y - yhat))

    loss = np.mean((y - yhat)**2)

    b = b - lr * grad_b
    w = w - lr * grad_w

    costs.append(loss)
    epochs.append(i)
  return w, b, costs, epochs
```
